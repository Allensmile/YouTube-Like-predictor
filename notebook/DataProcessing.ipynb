### This notebook contains data processing implementation for data cleaning and encoding catgorical fetaures

```python
import pandas as pd
import numpy as np
import pickle
import requests
import time
# pd.options.display.max_rows = 999
```

```python
df_new = pd.read_csv("dataset/data.csv")
df_new.columns
```

### Duration

```python
import isodate
df_new["duration"] = df_new.duration.apply(lambda x : isodate.parse_duration(x).total_seconds())
```

### Encoding categorical data

```python
cat_list = ['caption','projection','privacyStatus','license',"licencedContent","embeddable","dimension","definition",'categoryId','day_uploaded']
```

```python
for i in cat_list:
    print i
#     print df_new[i].unique()
    df_new[i] = df_new[i].astype('category')
    print dict( enumerate(df_new[i].cat.categories) )
```

```python
cat_columns = df_new.select_dtypes(['category']).columns
cat_columns
```

### Encoding dictionary for reference
caption
{0: False, 1: True}
projection
{0: '360', 1: 'rectangular'}
privacyStatus
{0: 'public', 1: 'unlisted'}
license
{0: 'creativeCommon', 1: 'youtube'}
licencedContent
{0: False, 1: True}
embeddable
{0: False, 1: True}
dimension
{0: '2d', 1: '3d'}
definition
{0: 'hd', 1: 'sd'}
categoryId
{0: 1, 1: 2, 2: 10, 3: 15, 4: 17, 5: 19, 6: 20, 7: 22, 8: 23, 9: 24, 10: 25, 11: 26, 12: 27, 13: 28, 14: 29}
day_uploaded
{0: 'Friday', 1: 'Monday', 2: 'Saturday', 3: 'Sunday', 4: 'Thursday', 5: 'Tuesday', 6: 'Wednesday'}

```python
df_new[cat_columns] = df_new[cat_columns].apply(lambda x: x.cat.codes)
```

```python
df_new
```

### Removing data wih no like counts

```python
like_count_null = list(df_new[df_new.likeCount.isnull()].index)
print len(like_count_null)
```

```python
df_new = df_new.drop(df_new.index[like_count_null]).reindex()
df_new.describe()
```

```python
df_new.to_csv("dataset/data.csv",header=True,index=False)
```

### Filling missing comment column with mean value

```python
df_new.commentCount = df_new.commentCount.fillna(df_new.commentCount.mean())
df_new.describe()
```
